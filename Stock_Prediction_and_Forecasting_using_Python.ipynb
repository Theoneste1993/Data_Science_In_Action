```python
stock=['GOOG','MSFT','AMZN','IBM']
df_yahoo=yf.download(stock, end='2022-12-01', start='2020-01-01')
```

```python
df_yahoo.iloc[:,:4].plot(figsize=(9,6))
import matplotlib.pyplot as plt
plt.legend()
plt.ylabel("Stock Price")
plt.xlabel("Time")
plt.title("Stock Price Against Time\n")
plt.show();
```

```python
Adj_close=df_yahoo['Adj Close']
Adj_close.head(12)
```

```python
# Rate of return =(Price_t-Price_t-1)/Price_t-1)
Rate_of_return=Adj_close.pct_change()
Rate_of_return.dropna(inplace=True)
```

```python
Rate_of_return.head()
```

```python
Rate_of_return.info()
```

```python
import matplotlib.pyplot as plt
Rate_of_return.plot(figsize=(9,6))
plt.legend()
plt.ylabel("Volatility of the stocks")
plt.xlabel('Time')
plt.title("Stock Price of the time\n")
plt.show();
```

```python
Rate_of_return.std()
```

*Simple Moving AVerage SMA of the Stock price*

```python
Rate_of_return[['Amz_60','GOOG_60','IBM_60','MSFT_60']]=Rate_of_return.rolling(120).mean()
```

```python
Rate_of_return.dropna(inplace=True)
Rate_of_return.head()
```

```python
Rolling_mean=Rate_of_return[['Amz_60','GOOG_60','IBM_60','MSFT_60']]
Rolling_mean.head()
```

```python
Rolling_mean.plot(figsize=(9,6))
```

```python
Rolling_mean.std()
```

*Letâ€™s import the augmented Dickey-Fuller test from the statsmodels package to test the stationarity of my stock price.*

```python
from statsmodels.tsa.stattools import adfuller
adft = adfuller(Rolling_mean['Amz_60'], autolag="AIC")
```

```python
output_df = pd.DataFrame({"Values":[adft[0],adft[1],adft[2],adft[3], 
                                    adft[4]['1%'], adft[4]['5%'], 
                                    adft[4]['10%']]  ,
  "Metric":["Test Statistics",
            "p-value",
            "No. of lags used",
            "Number of observations", 
            "critical value (1%)", 
            "critical value (5%)", 
            "critical value (10%)"]})
print(output_df)
```
